```python
class BOLLYWOOD_DATASET:
    def load_dataset(self, path, col):
        df = pd.read_csv(path, usecols=col)
        return df
    
    def get_shape(self, df):
        return df.shape
    
    def nanVal(self, df):
        null = df.isnull().sum()
        for nan in range(len(null)):
            if null[nan] > 0:
                df.dropna(inplace=True)
                print(f'{null[nan]} Nan Values Droped!')
            if null[nan] == 0:
                print("No Nan Values Found:)")
    
    def drop_col(self, df, col):
        df.drop(col, axis=1, inplace=True)
    
    def merge(self, left, right, on):
        return pd.merge(left=left, right=right, on=on)
    
    def rename(self, df, dict_data):
        df.rename(dict_data,axis=1,inplace=True)
    
    def replace(self, df, col, rp, to_rp):
        df[col] = df[col].str.replace(rp, to_rp)
    
    def getActors_bollywood(self, data,  director = False,  actor_1 = False,  actor_2 = False, actor_3 = False, actor_4 = False):
        data = str(data).split('|')
        try:
            if director == True:
                return data[0]
        except:
            return 'unknown'
        try:
            if actor_1 == True:
                return data[1]
        except:
            return 'unknown'
        try:
            if actor_2 == True:
                return data[2]
        except:
            return 'unknown'
        try:
            if actor_3 == True:
                return data[3]
        except:
            return 'unknown'
        try:
            if actor_4 == True:
                return data[4]
        except:
            return 'unknown'
    
    def combine(self, df):
        return df['director_name'] + ' ' + df['actor_1_name'] + ' ' + df['actor_2_name'] + ' ' + df['actor_3_name'] + ' ' + df['actor_4_name'] + ' ' + df['genres']

```

```python
bollywood_dataset = BOLLYWOOD_DATASET()
```

```python
df_1989_actors = bollywood_dataset.load_dataset(path='../input/the-indian-movie-database/1950-1989/bollywood_text_1950-1989.csv', col=['actors', 'imdb_id'])
df_1989_actors.head()
```

```python
df_1989_genres = bollywood_dataset.load_dataset(path='../input/the-indian-movie-database/1950-1989/bollywood_meta_1950-1989.csv', col=['genres', 'imdb_id', 'title'])
df_1989_genres.head()
```

```python
bollywood_dataset.get_shape(df_1989_genres), bollywood_dataset.get_shape(df_1989_actors)
```

```python
bollywood_dataset.nanVal(df_1989_genres), bollywood_dataset.nanVal(df_1989_actors)
```

```python
df_1989_new = bollywood_dataset.merge(left=df_1989_genres, right=df_1989_actors, on=['imdb_id'])
bollywood_dataset.drop_col(df_1989_new, ['imdb_id'])
bollywood_dataset.rename(df_1989_new, {'title':'movie_title'})
bollywood_dataset.replace(df_1989_new, 'genres', '|', " ")
df_1989_new.head()
df_1989_new.head()
```

```python
bollywood_dataset.nanVal(df_1989_new)
```

```python
df_1989_new['actors'][0]
```

```python
df_1989_new['director_name'] = df_1989_new['actors'].apply(lambda x: bollywood_dataset.getActors_bollywood(x, director=True))
df_1989_new['actor_1_name'] = df_1989_new['actors'].apply(lambda x: bollywood_dataset.getActors_bollywood(x, actor_1=True))
df_1989_new['actor_2_name'] = df_1989_new['actors'].apply(lambda x: bollywood_dataset.getActors_bollywood(x, actor_2=True))
df_1989_new['actor_3_name'] = df_1989_new['actors'].apply(lambda x: bollywood_dataset.getActors_bollywood(x, actor_3=True))
df_1989_new['actor_4_name'] = df_1989_new['actors'].apply(lambda x: bollywood_dataset.getActors_bollywood(x, actor_4=True))
bollywood_dataset.drop_col(df_1989_new,'actors')
df_1989_new['combination'] = bollywood_dataset.combine(df_1989_new)
df_1989_new.head()
```
